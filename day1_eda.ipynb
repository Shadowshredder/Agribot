```python
# day1_eda.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Step 1: Load the dataset
df = pd.read_csv("Crop_recommendation.csv")
```

```python
# Step 2: Preview the dataset
print(" First 5 rows of the dataset:")
print(df.head())
```

```python
# Step 3: Basic dataset info
print("\n Dataset Shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
print("\n Dataset Info:")
print(df.info())
```

```python
# Step 4: Check for missing/null values
print("\n Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Step 5: Statistical summary
print("\n Descriptive Statistics:")
print(df.describe())
```

```python
# Step 6: Unique crops
print("\n Unique Crops:")
unique_crops = df['label'].unique()
print(unique_crops)
print(f"Total number of crops: {len(unique_crops)}")
```

```python
# Step 7: Count of each crop (for balance checking)
plt.figure(figsize=(12, 6))
sns.countplot(y='label', data=df, order=df['label'].value_counts().index, palette="viridis")
plt.title("Crop Frequency Distribution")
plt.xlabel("Count")
plt.ylabel("Crop")
plt.tight_layout()
plt.show()
```

```python
# Step 8: Correlation heatmap (only numeric columns)
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # exclude 'label'
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# Step 9: Pairplot (commented to save time)
sns.pairplot(df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']])
plt.show()
```

```python
# Step 10: Save cleaned version
df.to_csv("clean_crop_data.csv", index=False)
print("\n Cleaned dataset saved as 'clean_crop_data.csv'")
```
